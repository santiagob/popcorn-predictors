```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ast #For visualizing variables stored in JSON format

# Load datasets
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')

print("Movies Dataframe:")
display(movies)
print("Credits Dataframe:")
display(credits)

# Display basic info about datasets
print("Movies Dataset:")
print(movies.info(), "\n")
print("Credits Dataset:")
print(credits.info(), "\n")

# Show the first few rows of both datasets
print(movies.head(), "\n")
print(credits.head(), "\n")

# Check for missing values
print("Missing values in Movies dataset:\n", movies.isnull().sum(), "\n")
print("Missing values in Credits dataset:\n", credits.isnull().sum(), "\n")

# Basic statistics of numerical columns
print("Movies dataset statistics:\n", movies.describe(), "\n")

# Visualizing distribution of movie ratings
plt.figure(figsize=(10, 5))
sns.histplot(movies['vote_average'].dropna(), bins=20, kde=True)
plt.xlabel("Average Vote")
plt.ylabel("Frequency")
plt.title("Distribution of Movie Ratings")
plt.show()

```

```python
# Rename 'movie_id' in credits to 'id' in credits to merge the two tables
credits.rename(columns={'movie_id': 'id'}, inplace=True)

# Merge the datasets on 'id'
movies = movies.merge(credits, on="id")

# Display available columns after merging
print("Columns in merged dataset:", movies.columns)

# Handle column renaming after merging (title is displayed twice)
if 'title' not in movies.columns:
    if 'title_x' in movies.columns:  # Keep the movie dataset title
        movies.rename(columns={'title_x': 'title'}, inplace=True)
    elif 'title_y' in movies.columns:
        movies.rename(columns={'title_y': 'title'}, inplace=True)

# --- Initial Feature Exploration ---

# Select relevant columns (I chose them  after looking into it a bit, we can change these if needed)
available_features = ['id', 'title','original_language', 'keywords', 'genres', 'budget', 'revenue','production_company', 'runtime', 
                      'popularity', 'vote_average', 'vote_count', 'release_date', 'crew', 'cast']

df_joined = movies[[col for col in available_features if col in movies.columns]].copy()

display(df_joined)

# Convert 'release_date' to datetime
df_joined['release_date'] = pd.to_datetime(df_joined['release_date'], errors='coerce')

# Extract release year
df_joined['release_year'] = df_joined['release_date'].dt.year

# Handle missing values ()
df_joined.fillna({'budget': 0, 'revenue': 0, 'runtime': df_joined['runtime'].median()}, inplace=True)

# Extract director from 'crew' column
def get_director(crew_data):
    try:
        if pd.isna(crew_data):
            return np.nan
        crew_list = ast.literal_eval(crew_data)
        for crew_member in crew_list:
            if crew_member.get('job') == 'Director':
                return crew_member['name']
    except (ValueError, SyntaxError, TypeError):
        return np.nan
    return np.nan

df_joined['director'] = df_joined['crew'].apply(get_director)

# Extract top 5 actors from 'cast' column
def get_main_actors(cast_data, top_n=5):
    try:
        if pd.isna(cast_data):
            return []
        cast_list = ast.literal_eval(cast_data)
        return [actor['name'] for actor in cast_list[:top_n]]
    except (ValueError, SyntaxError, TypeError):
        return []
    
df_joined['main_actors'] = df_joined['cast'].apply(get_main_actors)

# Drop 'crew' and 'cast' after extraction
df_joined.drop(columns=['crew', 'cast'], inplace=True, errors='ignore')

# Summary statistics
print("\nSummary Statistics:")
print(df_joined[['budget', 'revenue', 'runtime', 'popularity', 'vote_average', 'vote_count']].describe())

# Budget vs Revenue
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df_joined['budget'], y=df_joined['revenue'], alpha=0.5)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.title('Budget vs Revenue')
plt.show()

# Distribution of movie runtimes
plt.figure(figsize=(8, 5))
sns.histplot(df_joined['runtime'].dropna(), bins=30, kde=True)
plt.xlabel('Runtime (minutes)')
plt.title('Distribution of Movie Runtimes')
plt.show()

# Average vote distribution
plt.figure(figsize=(8, 5))
sns.histplot(df_joined['vote_average'], bins=20, kde=True)
plt.xlabel('Average Vote')
plt.title('Distribution of Vote Averages')
plt.show()

# Save cleaned dataset
df_joined.to_csv("cleaned_df_joined.csv", index=False)

# Display first few rows
print("\nCleaned dataset sample:")
print(df_joined[['id', 'title', 'release_year', 'budget', 'revenue', 'runtime', 'vote_average']].head())

# Summary of missing values
print("\nMissing values after joining and cleaning:")
print(df_joined.isnull().sum())

#Drop rows with null values
df_joined = df_joined.dropna().reset_index(drop=True)

# Summary of missing values after rows are dropped
print("\nMissing values after dropping rows with null values (Around 30, out of 4803)")
print(df_joined.isnull().sum())
```

```python
def plot_feature_analysis(df):
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # Top 10 Directors by Movie Count
    top_directors = df['director'].value_counts().nlargest(10)
    sns.barplot(y=top_directors.index, x=top_directors.values, ax=axes[0, 0], palette='magma')
    axes[0, 0].set_title("Top 10 Directors by Movie Count")
    
    # Top 10 Actors by Appearance
    actor_counts = pd.Series([actor for sublist in df['main_actors'] for actor in sublist]).value_counts()
    top_actors = actor_counts.nlargest(10)
    sns.barplot(y=top_actors.index, x=top_actors.values, ax=axes[0, 1], palette='coolwarm')
    axes[0, 1].set_title("Top 10 Actors by Movie Count")
    
    # Average Rating per Genre
    genre_ratings = {}
    for _, row in df.iterrows():
        for genre in row['genres']:
            if genre not in genre_ratings:
                genre_ratings[genre] = []
            genre_ratings[genre].append(row['vote_average'])
    avg_genre_ratings = {k: np.mean(v) for k, v in genre_ratings.items()}
    sorted_genres = sorted(avg_genre_ratings.items(), key=lambda x: x[1], reverse=True)
    sns.barplot(y=[g[0] for g in sorted_genres[:10]], x=[g[1] for g in sorted_genres[:10]], ax=axes[1, 0], palette='viridis')
    axes[1, 0].set_title("Top 10 Genres by Average Rating")
    
    # Popularity vs Revenue
    sns.scatterplot(x=df['popularity'], y=df['revenue'], ax=axes[1, 1], alpha=0.5)
    axes[1, 1].set_xscale("log")
    axes[1, 1].set_yscale("log")
    axes[1, 1].set_title("Popularity vs Revenue")
    axes[1, 1].set_xlabel("Popularity")
    axes[1, 1].set_ylabel("Revenue")
    
    plt.tight_layout()
    plt.show()

# Run the function
plot_feature_analysis(df_joined)
```

```python
# Merge the IMDB-style 'score' back to the main DataFrame
df = df_joined.merge(qualified[['title', 'score']], on='title', how='left')

# Fill missing scores with 0 since they don't have enough votes to qualify
df['score'] = df['score'].fillna(0)

#Correlation matrix

plt.figure(figsize=(10, 6))
sns.heatmap(df[['budget', 'revenue', 'runtime', 'popularity', 'vote_average', 'vote_count', 'release_year', 'score']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()


from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder

# Convert list-like strings to plain text
def process_list_column(col):
    return col.apply(lambda x: ' '.join(x) if isinstance(x, list) else str(x))

df['genres'] = process_list_column(df['genres'])
df['main_actors'] = process_list_column(df['main_actors'])
df['keywords'] = process_list_column(df['keywords'])

# TF-IDF Vectorization for textual features
tfidf = TfidfVectorizer(stop_words='english', max_features=300)

genres_matrix = tfidf.fit_transform(df['genres'])
actors_matrix = tfidf.fit_transform(df['main_actors'])
director_matrix = tfidf.fit_transform(df['director'])
keyword_matrix = tfidf.fit_transform(df['keywords'])

# One-Hot Encoding for original_language
ohe = OneHotEncoder(sparse_output=True)
language_matrix = ohe.fit_transform(df[['original_language']])


```

```python
from sklearn.preprocessing import StandardScaler
from scipy.sparse import hstack, csr_matrix
from sklearn.decomposition import PCA

# Choose numeric features
numeric_cols = ['budget', 'revenue','runtime','release_year', 'popularity', 'score']
numeric_data = df[numeric_cols].replace('', 0).astype(float)

#Normalize the data
scaled_numeric = StandardScaler().fit_transform(numeric_data)

# Combine all into one matrix including the text features (after one hot encoding and vectorization) and numeric features
combined_features = hstack([
    csr_matrix(scaled_numeric),
    genres_matrix,
    keyword_matrix,
    actors_matrix,
    director_matrix,
    language_matrix
])
# Use PCA on the dense version to find principal components that explain most of the variance in the model
pca = PCA(n_components=10)
pca_components = pca.fit_transform(combined_features.toarray())

feature_names = (
    numeric_cols +
    [f'genre_{i}' for i in range(genres_matrix.shape[1])] +
    [f'actor_{i}' for i in range(actors_matrix.shape[1])] +
    [f'director_{i}' for i in range(director_matrix.shape[1])] +
    [f'lang_{i}' for i in range(language_matrix.shape[1])]
)

for i, comp in enumerate(pca.components_):
    top_features = sorted(zip(comp, feature_names), key=lambda x: abs(x[0]), reverse=True)[:5]
    print(f"Component {i+1} top features:")
    for val, name in top_features:
        print(f"  {name}: {val:.3f}")
    print()

# Variance explained
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), pca.explained_variance_ratio_, marker='o')
plt.title('Explained Variance by PCA Components')
plt.xlabel('PCA Components')
plt.ylabel('Explained Variance Ratio')
plt.grid(True)
plt.show()

# Print component importance
for i, comp in enumerate(pca.components_):
    print(f"Component {i+1} explains {pca.explained_variance_ratio_[i]} of the variance")
print(f"Total variance explained by the model: {sum(pca.explained_variance_ratio_)}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances, manhattan_distances


# Train KNN models using the Principal Component Analysis results: 3 Models based on research on different methods KNNs use (and a combination of all 3)
# Based on some research, most movie recommendation systems use Cosine similarity, added Euclidian Distance and Manhattan Distances to compare results in the end
cos_sim = cosine_similarity(pca_components)
euc_sim = 1 / (1 + euclidean_distances(pca_components))
man_sim = 1 / (1 + manhattan_distances(pca_components))

# Average similarity model (can change weights as needed)
combined_sim = (cos_sim + euc_sim + man_sim) / 3
```

```python
#KNN models to recommend movies: 4 different models, Cosine Similarity, Euclidian Distance and Manhattan Distances

#Cosine similarity recommendation system
def recommend_movies_cosine(watched_titles, top_n=5):
    titles = df['title'].tolist()
    watched_indices = [i for i, t in enumerate(titles) if t in watched_titles]

    if not watched_indices:
        return "No matching titles found."

    sim_scores = cos_sim[watched_indices].mean(axis=0)
    for idx in watched_indices:
        sim_scores[idx] = -1  # Exclude watched

    recommended_indices = sim_scores.argsort()[-top_n:][::-1]
    return df.iloc[recommended_indices][['title', 'score', 'genres', 'main_actors','director','original_language']]

#Euclidian distance recommendation system
def recommend_movies_euclidian(watched_titles, top_n=5):
    titles = df['title'].tolist()
    watched_indices = [i for i, t in enumerate(titles) if t in watched_titles]

    if not watched_indices:
        return "No matching titles found."

    sim_scores = euc_sim[watched_indices].mean(axis=0)
    for idx in watched_indices:
        sim_scores[idx] = -1  # Exclude watched

    recommended_indices = sim_scores.argsort()[-top_n:][::-1]
    return df.iloc[recommended_indices][['title', 'score', 'genres', 'main_actors','director','original_language']]

#Manhattan distance recommendation system
def recommend_movies_manhattan(watched_titles, top_n=5):
    titles = df['title'].tolist()
    watched_indices = [i for i, t in enumerate(titles) if t in watched_titles]

    if not watched_indices:
        return "No matching titles found."

    sim_scores = man_sim[watched_indices].mean(axis=0)
    for idx in watched_indices:
        sim_scores[idx] = -1  # Exclude watched

    recommended_indices = sim_scores.argsort()[-top_n:][::-1]
    return df.iloc[recommended_indices][['title', 'score', 'genres', 'main_actors','director','original_language']]

#Combination of all models above
def recommend_movies(watched_titles, top_n=5):
    titles = df['title'].tolist()
    watched_indices = [i for i, t in enumerate(titles) if t in watched_titles]

    if not watched_indices:
        return "No matching titles found."

    sim_scores = combined_sim[watched_indices].mean(axis=0)
    for idx in watched_indices:
        sim_scores[idx] = -1  # Exclude watched

    recommended_indices = sim_scores.argsort()[-top_n:][::-1]
    return df.iloc[recommended_indices][['title', 'score', 'genres', 'main_actors','director','original_language']]
```

```python
import random
from sklearn.metrics import precision_score, recall_score, f1_score

# 3 Different functions to simulate users. Goal is to understand how accurate the recommendation systems are based on a relevance function (Completely subjective)
# Simulate watched lists for synthetic users
def simulate_users(df, num_users, watched_per_user):
    users = []
    for _ in range(num_users):
        watched = df.sample(watched_per_user)['title'].tolist()
        users.append(watched)
    return users

# Simulate watched lists for synthetic users but only movies with 7.0 scores or higher
def simulate_users_highscores(df, num_users, watched_per_user):
    users = []
    for _ in range(num_users):
        watched = df[df['score']>7.0].sample(watched_per_user)['title'].tolist()
        users.append(watched)
    return users

# Simulate watched lists for synthetic users but with only non-English movies
def simulate_users_non_English(df, num_users, watched_per_user):
    users = []
    for _ in range(num_users):
        watched = df[df['original_language']!='en'].sample(watched_per_user)['title'].tolist()
        users.append(watched)
    return users

# Relevance function only score and genre overlap
def is_relevant(movie, watched_scores, watched_genres):

    #If a movie in the recommended list has a score less than the average score minus 2 std deviations of scores
    # of the movies in the simulated set then it is considered not relevant
    score_threshold = watched_scores.mean() - 2 * watched_scores.std()

    if movie['score'] < score_threshold:
        return False

    # Compare genres using overlap ratio
    movie_genres = set(movie['genres'].replace('[{', '').replace('}]', '').replace('}, {', '|').split('|'))
    overlap_ratio = len(movie_genres.intersection(watched_genres)) / max(len(movie_genres), 1)

    return overlap_ratio >= 0.25  # Require at least 25% genre overlap

# Evaluation function
def evaluate_model(recommend_func, df, users, model_name="", n=None, scenario=""):
    precision_list = []
    recall_list = []
    f1_list = []
    average_precisions = []

    for watched_titles in users:
        watched_movies = df[df['title'].isin(watched_titles)]
        watched_genres = set('|'.join(watched_movies['genres'].astype(str)).replace('[{', '').replace('}]', '').replace('}, {', '|').split('|'))
        watched_scores = watched_movies['score']

        recommended = recommend_func(watched_titles)
        if isinstance(recommended, str) or recommended is None or recommended.empty:
            continue
        recommended_titles = recommended['title'].tolist()

        relevances = []
        for _, row in recommended.iterrows():
            relevant = is_relevant(row, watched_scores, watched_genres)
            relevances.append(1 if relevant else 0)

        tp = sum(relevances)
        fp = len(relevances) - tp
        all_relevant_items = df[df.apply(lambda row: is_relevant(row, watched_scores, watched_genres), axis=1)]
        fn = len(all_relevant_items) - tp

        precision = tp / (tp + fp) if (tp + fp) else 0
        recall = tp / (tp + fn) if (tp + fn) else 0
        f1 = (2 * precision * recall) / (precision + recall) if (precision + recall) else 0

        avg_prec = 0
        hit_count = 0
        for i, rel in enumerate(relevances):
            if rel:
                hit_count += 1
                avg_prec += hit_count / (i + 1)
        map_score = avg_prec / hit_count if hit_count else 0

        precision_list.append(precision)
        recall_list.append(recall)
        f1_list.append(f1)
        average_precisions.append(map_score)

    return {
        "Model": model_name,
        "Scenario": scenario,
        "n_movies": n,
        "Precision": sum(precision_list) / len(precision_list) if precision_list else 0,
        "Recall": sum(recall_list) / len(recall_list) if recall_list else 0,
        "F1": sum(f1_list) / len(f1_list) if f1_list else 0,
        "MAP": sum(average_precisions) / len(average_precisions) if average_precisions else 0,
    }

# Main loop to collect results
#Array for number of movies in the simulated movie sets
number_of_movies = [1, 5, 10, 20, 30]

#Create dictionaries to store all the simulated sets 
user_sets = {}
user_sets_highscore = {}
user_sets_non_English = {}

#Array to store results
all_results = []

#Model name and which recommendation function to use
models = [
    ("Cosine Similarity", recommend_movies_cosine),
    ("Euclidean Distance", recommend_movies_euclidian),
    ("Manhattan Distance", recommend_movies_manhattan)
]

#Simulating using 1000 users to ensure conversion
# Scenario 1: Generic: There are no restrictions, all movie sets are simulated using the entire set
for n in number_of_movies:
    users = simulate_users(df, 1000, n)
    user_sets[n] = users
    for model_name, model_func in models:
        result = evaluate_model(model_func, df, users, model_name=model_name, n=n, scenario="Generic")
        all_results.append(result)

# Scenario 2: High Scores Only: For this, movies with score greater than 7 are filtered to create the simulation sets
for n in number_of_movies:
    users_highscore = simulate_users_highscores(df, 1000, n)
    user_sets_highscore[n] = users_highscore
    for model_name, model_func in models:
        result = evaluate_model(model_func, df, users_highscore, model_name=model_name, n=n, scenario="High Score")
        all_results.append(result)

# Scenario 3: Non-English Only: For this, movies that don't have English as the main language are filtered to create the simulation sets
for n in number_of_movies:
    users_non_English = simulate_users_non_English(df, 1000, n)
    user_sets_non_English[n] = users_non_English
    for model_name, model_func in models:
        result = evaluate_model(model_func, df, users_non_English, model_name=model_name, n=n, scenario="Non-English")
        all_results.append(result)

# Create DataFrame
results_df = pd.DataFrame(all_results)
results_df.sort_values(by=['Model', 'Scenario'])
```
