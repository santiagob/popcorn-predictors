```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ast #For visualizing variables stored in JSON format

# Load datasets
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')

print("Movies Dataframe:")
display(movies)
print("Credits Dataframe:")
display(credits)

# Display basic info about datasets
print("Movies Dataset:")
print(movies.info(), "\n")
print("Credits Dataset:")
print(credits.info(), "\n")

# Show the first few rows of both datasets
print(movies.head(), "\n")
print(credits.head(), "\n")

# Check for missing values
print("Missing values in Movies dataset:\n", movies.isnull().sum(), "\n")
print("Missing values in Credits dataset:\n", credits.isnull().sum(), "\n")

# Basic statistics of numerical columns
print("Movies dataset statistics:\n", movies.describe(), "\n")

# Visualizing distribution of movie ratings
plt.figure(figsize=(10, 5))
sns.histplot(movies['vote_average'].dropna(), bins=20, kde=True)
plt.xlabel("Average Vote")
plt.ylabel("Frequency")
plt.title("Distribution of Movie Ratings")
plt.show()

```

```python
# Rename 'movie_id' in credits to 'id' in credits to merge the two tables
credits.rename(columns={'movie_id': 'id'}, inplace=True)

# Merge the datasets on 'id'
movies = movies.merge(credits, on="id")

# Display available columns after merging
print("Columns in merged dataset:", movies.columns)

# Handle column renaming after merging (title is displayed twice)
if 'title' not in movies.columns:
    if 'title_x' in movies.columns:  # Keep the movie dataset title
        movies.rename(columns={'title_x': 'title'}, inplace=True)
    elif 'title_y' in movies.columns:
        movies.rename(columns={'title_y': 'title'}, inplace=True)

# --- Initial Feature Exploration ---

# Select relevant columns (I chose them  after looking into it a bit, we can change these if needed)
available_features = ['id', 'title','original_language', 'genres', 'budget', 'revenue','production_company', 'runtime', 
                      'popularity', 'vote_average', 'vote_count', 'release_date', 'crew', 'cast']

df_joined = movies[[col for col in available_features if col in movies.columns]].copy()

display(df_joined)

# Convert 'release_date' to datetime
df_joined['release_date'] = pd.to_datetime(df_joined['release_date'], errors='coerce')

# Extract release year
df_joined['release_year'] = df_joined['release_date'].dt.year

# Handle missing values ()
df_joined.fillna({'budget': 0, 'revenue': 0, 'runtime': df_joined['runtime'].median()}, inplace=True)

# Extract director from 'crew' column
def get_director(crew_data):
    try:
        if pd.isna(crew_data):
            return np.nan
        crew_list = ast.literal_eval(crew_data)
        for crew_member in crew_list:
            if crew_member.get('job') == 'Director':
                return crew_member['name']
    except (ValueError, SyntaxError, TypeError):
        return np.nan
    return np.nan

df_joined['director'] = df_joined['crew'].apply(get_director)

# Extract top 5 actors from 'cast' column
def get_main_actors(cast_data, top_n=5):
    try:
        if pd.isna(cast_data):
            return []
        cast_list = ast.literal_eval(cast_data)
        return [actor['name'] for actor in cast_list[:top_n]]
    except (ValueError, SyntaxError, TypeError):
        return []
    
df_joined['main_actors'] = df_joined['cast'].apply(get_main_actors)

# Drop 'crew' and 'cast' after extraction
df_joined.drop(columns=['crew', 'cast'], inplace=True, errors='ignore')

# Summary statistics
print("\nSummary Statistics:")
print(df_joined[['budget', 'revenue', 'runtime', 'popularity', 'vote_average', 'vote_count']].describe())

# Budget vs Revenue
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df_joined['budget'], y=df_joined['revenue'], alpha=0.5)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.title('Budget vs Revenue')
plt.show()

# Distribution of movie runtimes
plt.figure(figsize=(8, 5))
sns.histplot(df_joined['runtime'].dropna(), bins=30, kde=True)
plt.xlabel('Runtime (minutes)')
plt.title('Distribution of Movie Runtimes')
plt.show()

# Average vote distribution
plt.figure(figsize=(8, 5))
sns.histplot(df_joined['vote_average'], bins=20, kde=True)
plt.xlabel('Average Vote')
plt.title('Distribution of Vote Averages')
plt.show()

# Save cleaned dataset
df_joined.to_csv("cleaned_df_joined.csv", index=False)

# Display first few rows
print("\nCleaned dataset sample:")
print(df_joined[['id', 'title', 'release_year', 'budget', 'revenue', 'runtime', 'vote_average']].head())

# Summary of missing values
print("\nMissing values after joining and cleaning:")
print(df_joined.isnull().sum())

#Drop rows with null values
df_joined = df_joined.dropna().reset_index(drop=True)

# Summary of missing values after rows are dropped
print("\nMissing values after dropping rows with null values (Around 30, out of 4803)")
print(df_joined.isnull().sum())
```

```python
# Extract genre list
def extract_genres(genre_str):
    if isinstance(genre_str, str):  # Only apply ast.literal_eval if it's a string
        genres = ast.literal_eval(genre_str)  # Convert string to list of dictionaries
        return [g['name'] for g in genres]
    elif isinstance(genre_str, list):  # Already a list, return as is
        return genre_str
    else:
        return []  # Handle potential NaN or other unexpected types

df_joined['genres'] = df_joined['genres'].apply(extract_genres)
genre_list = []
for genres in df_joined['genres']:
    genre_list.extend(genres)  # Ensure it's properly flattened

# Get unique genres and counts manually
unique_genres = list(set(genre_list))
genre_counts = [genre_list.count(genre) for genre in unique_genres]

# Plot most common genres
plt.figure(figsize=(12, 6))
sns.barplot(x=unique_genres, y=genre_counts, palette="viridis")
plt.xticks(rotation=45)
plt.xlabel("Genres")
plt.ylabel("Number of Movies")
plt.title("Most Common Movie Genres")
plt.show()
genre_list = []
for genres in df_joined['genres']:
    genre_list.extend(genres)  # Flatten list

# Original Language plots
unique_languages = df_joined['original_language'].unique()
language_counts = [sum(df_joined['original_language'] == lang) for lang in unique_languages]

# All Language plot 
plt.figure(figsize=(10, 5))
plt.bar(unique_languages, language_counts)
plt.ylabel("Number of Movies")
plt.xlabel("Original Language")
plt.title("Distribution of Movies by Original Language")
plt.xticks(rotation=45)  # Rotate labels for readability
plt.show()

# Filter out 'en' from unique languages and counts
filtered_languages = [lang for lang in unique_languages if lang != 'en']
filtered_counts = [sum(df_joined['original_language'] == lang) for lang in filtered_languages]

# Language Plots w/out English since it is dominating
plt.figure(figsize=(10, 5))
plt.bar(filtered_languages, filtered_counts, color=plt.cm.viridis(np.linspace(0, 1, len(filtered_languages))))
plt.ylabel("Number of Movies")
plt.xlabel("Original Language")
plt.title("Distribution of Movies by Original Language (Excluding English)")
plt.xticks(rotation=45)  # Rotate labels for readability
plt.show()
print(len(filtered_languages))
print(sum(filtered_counts))

# Extract years and genres
years = df_joined['release_year'].dropna().astype(int).unique()
years.sort()

# Prepare a dictionary to count genre occurrences per year
genre_year_dict = {year: {} for year in years}

# Count genres manually
for _, row in df_joined.iterrows():
    year = row['release_year']
    if not np.isnan(year):
        year = int(year)
        for genre in row['genres']:
            if genre in genre_year_dict[year]:
                genre_year_dict[year][genre] += 1
            else:
                genre_year_dict[year][genre] = 1

# Get all unique genres
all_genres = set()
for year_data in genre_year_dict.values():
    all_genres.update(year_data.keys())

# Plot
plt.figure(figsize=(14, 7))
for genre in all_genres:
    genre_counts = [genre_year_dict[year].get(genre, 0) for year in years]
    plt.plot(years, genre_counts, label=genre)

plt.xlabel("Release Year")
plt.ylabel("Number of Movies")
plt.title("Most Popular Genres Over Time")
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.show()

```

```python
def plot_feature_analysis(df):
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # Top 10 Directors by Movie Count
    top_directors = df['director'].value_counts().nlargest(10)
    sns.barplot(y=top_directors.index, x=top_directors.values, ax=axes[0, 0], palette='magma')
    axes[0, 0].set_title("Top 10 Directors by Movie Count")
    
    # Top 10 Actors by Appearance
    actor_counts = pd.Series([actor for sublist in df['main_actors'] for actor in sublist]).value_counts()
    top_actors = actor_counts.nlargest(10)
    sns.barplot(y=top_actors.index, x=top_actors.values, ax=axes[0, 1], palette='coolwarm')
    axes[0, 1].set_title("Top 10 Actors by Movie Count")
    
    # Average Rating per Genre
    genre_ratings = {}
    for _, row in df.iterrows():
        for genre in row['genres']:
            if genre not in genre_ratings:
                genre_ratings[genre] = []
            genre_ratings[genre].append(row['vote_average'])
    avg_genre_ratings = {k: np.mean(v) for k, v in genre_ratings.items()}
    sorted_genres = sorted(avg_genre_ratings.items(), key=lambda x: x[1], reverse=True)
    sns.barplot(y=[g[0] for g in sorted_genres[:10]], x=[g[1] for g in sorted_genres[:10]], ax=axes[1, 0], palette='viridis')
    axes[1, 0].set_title("Top 10 Genres by Average Rating")
    
    # Popularity vs Revenue
    sns.scatterplot(x=df['popularity'], y=df['revenue'], ax=axes[1, 1], alpha=0.5)
    axes[1, 1].set_xscale("log")
    axes[1, 1].set_yscale("log")
    axes[1, 1].set_title("Popularity vs Revenue")
    axes[1, 1].set_xlabel("Popularity")
    axes[1, 1].set_ylabel("Revenue")
    
    plt.tight_layout()
    plt.show()

# Run the function
plot_feature_analysis(df_joined)
```
